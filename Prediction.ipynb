```python
import pandas as pd
import pickle
```

```python
test = pd.read_csv('dataset_titanic/test.csv')
test.head()
```

```python
# Jika Wanita selamat dan semua anak anak selamat

def umur(x):
    if x <= 12:
        return "1"
    
    elif x >= 13 and x <= 18:
        return "2"
    
    elif x >= 19 and x <= 40:
        return "3"
    
    elif x >= 41 and x <= 65:
        return "4"    

    else:
        return "5"
    
test['umur'] = test['Age'].apply(lambda x : umur(x))
test['WA'] = 0
test.loc[(test['umur'] == '1') | (test['Sex'] == 'female'), 'WA'] = 1
test.head()
```

```python
test = test.drop(['Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1)
test
```

```python
model = pickle.load(open('model.sav', 'rb'))
```

```python
y_pred = model.predict(test)
```

```python
test['Survived'] = y_pred
test
```

```python
test = pd.read_csv('dataset_titanic/test.csv')
test['Survived'] = y_pred
test.head()
```

```python
submisi = test[['PassengerId', 'Survived']]
submisi
```

```python
submisi.to_csv('submisi.csv', index = False)
```
